```julia
function ac_ex_visualize(after_path)
    df = CSV.read(path * after_path, DataFrame)
    ac_ex = Matrix(df[:, :])

#採択率・交換率プロット
plot(ac_ex[:,1], st=:scatter, label="B20",alpha=0.5)
plot!(ac_ex[:,2], st=:scatter, label="B40",alpha=0.5, marker=:rect)
plot!(ac_ex[:,3], st=:scatter, label="B44",alpha=0.5, marker=:utriangle)
plot!(ac_ex[1:end-1,4], st=:scatter, label="ex_rate",alpha=0.5, marker=:diamond)

plot!(xlabel="replica", ylabel="rate")
plot!(legend=:topright)

end
```
